```python
ctn_sites = Counter(dff['Site'])
dict_ctn_sites = dict(ctn_sites)

df_ctn_sites = pd.DataFrame(dict_ctn_sites.items(), columns=['Site', 'Occurance'])
```

```python
import plotly.express as px

pie = px.pie(df_ctn_sites, values='Occurance', names='Site', title='counts of sites', width=1000, height=800)
pie.show()
```
